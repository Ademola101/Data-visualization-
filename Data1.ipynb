```python
import csv
filename = 'sitka_weather_07-2014.csv'
with open(filename) as f :
    reader = csv.reader(f)
    header_row = next(reader)
    #print header row
    high = []
    for row in reader:
        highs = int(row[1])
        high.append(highs)
#print high
from matplotlib import pyplot as plt
#plot data
fig =plt.figure(dpi = 128, figsize = (10,6))
plt.plot(high, c="red")
plt.title("Daily high Temperature of the month July", fontsize = 16)
plt.xlabel("",fontsize = 16)
plt.ylabel("Temperature F", fontsize = 16)
plt.tick_params(axis = "both", which = "major", labelsize = 16)
plt.show
```

```python
import csv
#import datetime module
from datetime import datetime
filename = 'sitka_weather_2014.csv'
with open(filename) as f :
    reader = csv.reader(f)
    header_row = next(reader)
    #print header row
    high = []
    date = []
    for row in reader:
        highs = int(row[1])
        high.append(highs)
        current_date = datetime.strptime(row[0],"%Y-%m-%d")
        date.append(current_date)
#print high
from matplotlib import pyplot as plt
#plot data
fig =plt.figure(dpi = 128, figsize = (10,6))
plt.plot(date,high, c="red")
plt.title("Daily high Temperature of the month July", fontsize = 16)
plt.xlabel("",fontsize = 16)
fig.autofmt_xdate()
plt.ylabel("Temperature F", fontsize = 16)
plt.tick_params(axis = "both", which = "major", labelsize = 16)
plt.show
```

```python
import csv
#import datetime module
from datetime import datetime
filename = 'sitka_weather_2014.csv'
with open(filename) as f :
    reader = csv.reader(f)
    header_row = next(reader)
    #print header row
    high = []
    low = []
    date = []
    for row in reader:
        highs = int(row[1])
        high.append(highs)
        current_date = datetime.strptime(row[0],"%Y-%m-%d")
        date.append(current_date)
        lows = int(row[3])
        low.append(lows)
#print high
from matplotlib import pyplot as plt
#plot data
fig =plt.figure(dpi = 128, figsize = (10,6))
plt.plot(date,low, c = "blue")
plt.plot(date,high, c="red")
plt.title("Daily high Temperature and low of the month July", fontsize = 16)
plt.xlabel("",fontsize = 16)
fig.autofmt_xdate()
plt.ylabel("Temperature F", fontsize = 16)
plt.tick_params(axis = "both", which = "major", labelsize = 16)
plt.show
```
